```python
#выравнивание длин - обрезка всех файлов по уровню самого короткого файла
import pandas as pd
import numpy as np
import random
import os
import soundfile as sf
import librosa
from pathlib import Path

#узнаем минимальный размер в папках из types
rootdir = os.getcwd()
types = ['test', 'train']
min_sizes = []
for type_ in types:
    sizes = []
    for root, dirs, files in os.walk(rootdir):
        for file in files:
            filepath = os.path.join(root, file)
            if Path(file).suffix in ['.wav'] and root == os.path.join(rootdir, type_):
                with sf.SoundFile(filepath, 'r', closefd = True) as f:
                    x = f.read()
                    sr = f.samplerate
                    channels = f.channels
                    f.close()
                if len(x) != 0:
                    sizes.append([file, len(x)])
    size_df = pd.DataFrame(sizes, columns = ['file', 'size'])
    print(size_df['size'])
    min_ind = size_df['size'].idxmin()
    min_size = size_df.iloc[min_ind]
    min_sizes.append(min_size['size'])
    print(size_df)
    print(f'Минимальный размер для {type_}: ', size_df.iloc[min_ind])

#обрезаем файлы по минимальному
rootdir = os.getcwd()
for type_, size in zip(types, min_sizes):
    for root, dirs, files in os.walk(os.path.join(rootdir, type_)):
        for file in files:
            filepath = os.path.join(root, file)
            if Path(file).suffix in ['.wav']:
                with sf.SoundFile(filepath, 'r', closefd = True) as f:
                    x = f.read()
                    sr = f.samplerate
                    channels = f.channels
                    f.close()
                x = x[: size]
                with sf.SoundFile(filepath, 'w', samplerate = sr, channels = channels) as f:
                    f.write(x)
                    f.close()
```
